```python
"""
Characteristics of this dataset:

This data set contains various data on all NBA players in the 2023-2024 NBA regular season. The player data it contains includes Player's name, Position, Player's age, Player's
  Team, Games played, Games started, Minutes played per game, various shooting data, shooting efficiency, rebound data, steal data and block data, etc. This is everything in the header of the data table and what it means:

Rk: Rank
Player : Player's name
Pos: Position
Age: Player's age
Tm:Team
G : Games played
GS: Games started
MP: Minutes played per game
FG: Field goals per game
FGA: Field goal attempts per game
FG%: Field goal percentage
3P: 3-point field goals per game
3PA: 3-point field goal attempts per game
3P%: 3-point field goal percentage
2P: 2-point field goals per game
2PA: 2-point field goal attempts per game
2P%: 2-point field goal percentage
eFG% : Effective field goal percentage
FT: Free throws per game
FTA: Free throw attempts per game
FT%: Free throw percentage
ORB: Offensive rebounds per game
DRB: Defensive rebounds per game
TRB : Total rebounds per game
AST: Assists per game
STL: Steals per game
BLK : Blocks per game
TOV: Turnovers per game
PF : Personal fouls per game
PTS: Points per game

The data set records a total of 735 pieces of data. It should be noted that some players will change teams during the season. Therefore, a player will have multiple pieces of data on different teams at the same time.
"""
```

```python
# Try loading the data with a different encoding, commonly 'ISO-8859-1' for CSV files with non-standard characters
data_iso = pd.read_csv("data/2023-2024 NBA Player Stats - Regular.csv", delimiter=";", encoding='ISO-8859-1')

# Display the first few rows of the dataframe
data_iso.head()
```

a pie chart illustrating the age distribution of NBA players for the 2023-2024 season, categorized into four age ranges:

Under 22 years old
22 to 28 years old
28 to 34 years old
Above 34 years old

```python
# Categorize the players based on age ranges
age_categories = pd.cut(data_iso['Age'], bins=[0, 21, 28, 34, 100], right=False, labels=["Under 22", "22 to 28", "28 to 34", "Above 34"])

# Count the number of players in each age category
age_distribution = age_categories.value_counts()

# Create a pie chart for the age distribution
plt.figure(figsize=(8, 8))
age_distribution.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange', 'red'])
plt.title('Age Distribution of NBA Players (2023-2024 Season)')
plt.ylabel('')  # Hide the y-label as it's not needed for pie charts
plt.show()
```

The bar chart above displays the players with the most backcourt rebounds (also referred to as offensive rebounds) by team for the 2023-2024 NBA season. Each bar represents a player, with their team in parentheses, and the length of the bar indicates the number of backcourt rebounds they accumulated, leading their respective teams in this category.

```python
# Filter the columns needed
player_rebounds = data_iso[['Player', 'Tm', 'ORB']]

# Group by team and find the player with the maximum offensive (backcourt) rebounds in each team
max_rebounds_by_team = player_rebounds.groupby('Tm').apply(lambda x: x.loc[x['ORB'].idxmax()])

# Sort by backcourt rebounds for better visualization
max_rebounds_sorted = max_rebounds_by_team.sort_values('ORB', ascending=False)

# Create a bar chart for the players with the most backcourt rebounds in each team
plt.figure(figsize=(12, 8))
plt.barh(max_rebounds_sorted['Player'] + " (" + max_rebounds_sorted['Tm'] + ")", max_rebounds_sorted['ORB'], color='purple')
plt.xlabel('Backcourt Rebounds')
plt.title('Players with the Most Backcourt Rebounds by Team (2023-2024 Season)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

 pie chart displaying the distribution of players based on their offensive rebounds per game for the 2023-2024 NBA season, with the new specified ranges:

0 to 1 rebounds
1 to 2 rebounds
2 to 3 rebounds
3 to 4 rebounds
More than 4 rebounds

```python
# Adjust the categories for the offensive rebounds as specified
orb_categories_adjusted = pd.cut(data_iso['ORB'], bins=[0, 1, 2, 3, 4, float('inf')], right=False, labels=["0 to 1", "1 to 2", "2 to 3", "3 to 4", "More than 4"])

# Count the number of players in each adjusted offensive rebound category
orb_distribution_adjusted = orb_categories_adjusted.value_counts()

# Create a pie chart for the adjusted offensive rebound distribution
plt.figure(figsize=(8, 8))
orb_distribution_adjusted.plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'orange', 'red', 'purple'])
plt.title('Distribution of Players by Offensive Rebounds (2023-2024 Season) [Adjusted Ranges]')
plt.ylabel('')  # Hide the y-label as it's not needed for pie charts
plt.show()
```
